# Step 2 : Load the Data

```python
df = pd.read_csv("../Dataset/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
```

# Feature Engineering

### Creating New Feautres

```python
df['tenure'].min(), df['tenure'].max()
df['tenure']
```

```python
df['tenure_group'] = pd.cut(df['tenure'],
    bins = [-1, 12, 24, 48, 73], 
    labels = [0, 1, 2, 3], 
    include_lowest = True)
```

```python
df[['tenure', 'tenure_group']]
```

```python
df.info()
```

```python
df['tenure_group'] = df['tenure_group'].astype(float).astype(int)
df['tenure_group'].info()
```

```python
df['avg_monthly_per_tenure'] = df['TotalCharges'] / (df['tenure'] + 1)
```

```python
df[df['avg_monthly_per_tenure'] == 'np.inf']
```

```python
df.columns
```

## Encoding

```python
df = df.drop('customerID', axis=1)
```

```python
df.head(2)
```

```python
binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
for col in binary_cols:
    df[col] = df[col].map({'Yes':1, 'No':0})

df.head(1)
```

```python
df['Churn'] = df['Churn'].map({"Yes":1, "No":0})
```

```python
df['gender'] = df['gender'].map({"Yes":1, "No":0})
```

```python
df.info()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
```

```python
for col in categorical_cols:
    le = LabelEncoder()

    df[col] = le.fit_transform(df[col].astype(str))
```
