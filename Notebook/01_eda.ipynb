### Visualise Target Variable

```python
fig, ax = plt.subplots(1,2, figsize=(14, 5))

# First Graph
sns.countplot(data=df, x='Churn', ax=ax[0], palette='Set2')
ax[0].set_title('Churn Distribution', fontweight='bold')
ax[0].set_ylabel('count')


# second graph
churn_count = df['Churn'].value_counts()
ax[1].pie(churn_count,
          labels= churn_count.index,
          autopct = '%1.1f%%',
          startangle = 90,
          colors=['#21d830', '#ce5050']
          )
ax[1].set_title('Churn Percentage', fontweight='bold')


plt.tight_layout()

plt.savefig('../Report/churn_distribution.png', dpi= 300, bbox_inches='tight')
plt.show()

print(f"Image saved : churn_distribution.png in Report folder")
```

### Numerical Features VS Churn

### Analyse Categorical Features

```python
categorical_col = df.select_dtypes(include=['object']).columns.tolist()
categorical_col
```

```python
categorical_col.remove('customerID')
categorical_col.remove('Churn')
categorical_col
```

```python
print(f"Categorical Columns : ({len(categorical_col)}): {categorical_col}")
```

```python
fig, ax = plt.subplots(5, 3, figsize=(20, 30))
ax = ax.flatten()

for idx, col in enumerate(categorical_col):
    churn_rate = df.groupby(col)['Churn'].apply(lambda x: (x=='Yes').sum()/len(x)*100)
    churn_rate.plot(kind='bar', ax = ax[idx], color="coral")
    ax[idx].set_title(f"Churn Rate by {col}, ", fontweight='bold')
    ax[idx].set_ylabel('Churn Rate %')
    ax[idx].set_xlabel(col)
    ax[idx].tick_params(axis='x', rotation=45)

plt.title('Categorical Features v/s Churn Rate')
plt.tight_layout()

plt.savefig('../Report/Categorical_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
df_corr = df.copy()
df_corr['Churn'] = df_corr['Churn'].map({'Yes': 1, 'No': 0})
```

```python
numerical_fetures = ['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']
corr_matrix = df_corr[numerical_fetures].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, square=True, linewidths=1)
plt.title("feature correlation Matrix heatmap", fontweight='bold')


plt.savefig('../Report/correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
```
